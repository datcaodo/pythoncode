## Question 4 - Creating the loyalty flag for existing customers

```python
#Splitting the data in 'user_id'
ords_prods_merge.groupby('user_id')
```

```python
#Creating a new column, 'max_order'
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

We do not have almost 16 million (15876776) Regular customers, almost 10 million (10284093) Loyal customers, and almost 6 million (6243990) New customers.

```python
# That's why we should count unique (Distinct) values of user_id in each group: 

ords_prods_merge.groupby(['loyalty_flag']).user_id.nunique() 

# here each user is counted only once
```

## Question 5 - Determine whether the prices of products purchased by local custers differ from those purchased by regular or new customers

```python
# Use groupby and agg() to determine basic statisitcs of spending habits by loyalty group
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

The average price of a new customer is about 3 dollars more than of a loyal customer. 

## Question 6 - Creating a spending flag 

```python
#Use transform() to determine the mean price of products purchased by each user
ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# View relevant columns 
ords_prods_merge[['user_id', 'mean_product_price']].head(20)
```

```python
#Create a new spending_flag column
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
#Check frequency of new spending_flag column
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# View relevant columns of updated data frame
ords_prods_merge[['user_id', 'mean_product_price', 'spending_flag']].head(20)
```

## Question 7 - Creating an order frequency flag 

```python
# Use groupby and transform to determine median number of days between orders for each user
ords_prods_merge['median_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

## Create a new frequency_flag column

```python
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] > 20, 'frequency_flag'] = 'Non-frequent Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] > 10) & (ords_prods_merge['median_days_between_orders']<= 20) , 'frequency_flag'] = 'Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
# Check value counts of new order_frequency column
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# View relevant columns of updated data frame
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders', 'frequency_flag']].head(25)
```
