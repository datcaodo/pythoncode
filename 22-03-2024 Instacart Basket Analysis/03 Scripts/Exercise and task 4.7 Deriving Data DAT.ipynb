### Defining a filter through the if, then function

```python
#visual hierarchy will help you determine what code belongs with what
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-rage product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else: return 'Not enough data'
```

If the value in the “prices” column within the given row is less than or equal to 5, then return the string “Low-range product.”

Or else, if the value in the “prices” column within the given row is greater than 5 and less than or equal to 15, then return the string “Mid-range product.”

Or else, if the value of the “prices” column within the given row is greater than 15, then return the string “High-range product.”

Or else, return the string “Not enough data.”

```python
#Using the function
df['price_range'] = df.apply(price_label, axis = 1)

# This creates a new column within the df dataframe called “price_range” 
#and designates it as the location for your labels.
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

#### Use for-loop to determine whether orders are on "busiest day" (0 = Saturday), "least busy" (4 = Wednesday), or "regularly busy" (other days of the week).


```python
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create new column in df_ords_prods_merge to view results in context.

df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna=False)
```
